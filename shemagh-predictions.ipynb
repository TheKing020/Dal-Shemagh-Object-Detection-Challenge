```python
import pandas as pd
```

```python
sub = pd.read_csv('/kaggle/input/notebooks/kagglertw/shemagh-simple-ddq-map-50-o-inference5/submission.csv')
```

```python
rp = pd.read_csv('/kaggle/input/notebooks/kagglertw/shemagh-right-place-binary-inference/predictions.csv')
```

```python
rp
```

```python
rp['prediction_class'] = (rp[['fold_1_prob', 'fold_2_prob', 'fold_3_prob', 'fold_4_prob', 'fold_5_prob']].mean(axis=1) > 0.653).astype(int)
```

```python
for _, row in rp.iterrows():
    sub.loc[
        (sub['filename'] == row['filename']) & (sub['right_place'] == 0),
        'right_place'
    ] = row['prediction_class']
```

```python
sub.to_csv('testing_.csv' , index = False)
```
