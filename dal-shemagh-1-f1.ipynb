```python
import glob
import os
import cv2

SRC_LABELS = "/kaggle/input/datasets/kagglertw/dal-shemagh/dal-shemagh-detection-challenge/labels/train"

def convert_annotations(image_folder, output_label_folder):

    os.makedirs(output_label_folder, exist_ok=True)

    for img_path in glob.glob(f"{image_folder}/*.jpg"):

        name = os.path.splitext(os.path.basename(img_path))[0]
        label_path = f"{SRC_LABELS}/{name}.txt"

        if not os.path.exists(label_path):
            continue

        img = cv2.imread(img_path)
        h, w = img.shape[:2]

        new_lines = []

        with open(label_path, "r") as f:
            for line in f:

                cls, x, y, bw, bh = map(float, line.split())


                x_center = x * w
                y_center = y * h
                bw_px = bw * w
                bh_px = bh * h


                x_min = x_center - bw_px / 2
                y_min = y_center - bh_px / 2

                new_lines.append(f"{int(cls)} {x_min} {y_min} {bw_px} {bh_px}\n")

        out_path = f"{output_label_folder}/{name}.txt"
        with open(out_path, "w") as f:
            f.writelines(new_lines)



convert_annotations(
    "/kaggle/input/notebooks/kagglertw/data-management-5f/data_5fold/fold_1/val",
    "/kaggle/working/val/labels"
)

convert_annotations(
    "/kaggle/input/notebooks/kagglertw/data-management-5f/data_5fold/fold_1/train",
    "/kaggle/working/train/labels"
)
```
