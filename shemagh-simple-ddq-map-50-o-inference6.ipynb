```python
%%capture
import pandas as pd
import numpy as np


result_df['xmin'] = pd.to_numeric(result_df['xmin'], errors='coerce')
result_df['ymin'] = pd.to_numeric(result_df['ymin'], errors='coerce')
result_df['xmax'] = pd.to_numeric(result_df['xmax'], errors='coerce')
result_df['ymax'] = pd.to_numeric(result_df['ymax'], errors='coerce')


def get_box_center(row):
    x_center = (row['xmin'] + row['xmax']) / 2
    y_center = (row['ymin'] + row['ymax']) / 2
    return x_center, y_center


def calculate_distance(center1, center2):
    return np.sqrt((center1[0] - center2[0])**2 + (center1[1] - center2[1])**2)


def calculate_iou(box1, box2):
    
    
    
    x_left = max(box1['xmin'], box2['xmin'])
    y_top = max(box1['ymin'], box2['ymin'])
    x_right = min(box1['xmax'], box2['xmax'])
    y_bottom = min(box1['ymax'], box2['ymax'])
    
    
    if x_right < x_left or y_bottom < y_top:
        return 0.0
    
    
    intersection_area = (x_right - x_left) * (y_bottom - y_top)
    
    
    box1_area = (box1['xmax'] - box1['xmin']) * (box1['ymax'] - box1['ymin'])
    box2_area = (box2['xmax'] - box2['xmin']) * (box2['ymax'] - box2['ymin'])
    
    
    union_area = box1_area + box2_area - intersection_area
    
    
    iou = intersection_area / union_area if union_area > 0 else 0.0
    
    return iou


image_ids_with_both = []
grouped = result_df.groupby('Image_ID')

for image_id, group in grouped:
    
    valid_group = group[group['class'].notna()]
    classes = valid_group['class'].unique()
    
    
    if 0.0 in classes and 1.0 in classes:
        image_ids_with_both.append(image_id)

print(f"Image IDs with both classes: {image_ids_with_both}")


result_dict = {}

for image_id in image_ids_with_both:
    group = result_df[result_df['Image_ID'] == image_id]
    
    
    class_1_boxes = group[(group['class'] == 1.0) & 
                          (group['xmin'].notna()) & 
                          (group['xmax'].notna())].copy()
    
    class_0_boxes = group[(group['class'] == 0.0) & 
                          (group['xmin'].notna()) & 
                          (group['xmax'].notna())].copy()
    
    if len(class_1_boxes) == 0 or len(class_0_boxes) == 0:
        continue
    
    
    class_1_row = class_1_boxes.iloc[0]
    class_1_center = get_box_center(class_1_row)
    
    
    min_distance = float('inf')
    closest_class_0_row = None
    
    for idx, class_0_row in class_0_boxes.iterrows():
        class_0_center = get_box_center(class_0_row)
        distance = calculate_distance(class_1_center, class_0_center)
        
        if distance < min_distance:
            min_distance = distance
            closest_class_0_row = class_0_row
    
    
    result_dict[image_id] = {
        'class_1_coords': {
            'xmin': class_1_row['xmin'],
            'ymin': class_1_row['ymin'],
            'xmax': class_1_row['xmax'],
            'ymax': class_1_row['ymax']
        },
        'class_0_coords': {
            'xmin': closest_class_0_row['xmin'],
            'ymin': closest_class_0_row['ymin'],
            'xmax': closest_class_0_row['xmax'],
            'ymax': closest_class_0_row['ymax']
        },
        'class_1_center': class_1_center,
        'class_0_center': get_box_center(closest_class_0_row)
    }

print(f"\nResult dictionary: {result_dict}")


summary_data = []

for image_id, data in result_dict.items():
    class_1_center_y = data['class_1_center'][1]
    class_0_center_y = data['class_0_center'][1]
    
    
    class_1_upper = data['class_1_coords']['ymin']
    class_0_upper = data['class_0_coords']['ymin']
    
    
    iou = calculate_iou(data['class_1_coords'], data['class_0_coords'])
    
    
    
    is_above = class_0_center_y < class_1_center_y
    
    
    
    if class_1_upper < class_0_upper and iou > 0.005:
        is_above = False
    
    summary_data.append({
        'Image_ID': image_id,
        'class_1_xmin': data['class_1_coords']['xmin'],
        'class_1_ymin': data['class_1_coords']['ymin'],
        'class_1_xmax': data['class_1_coords']['xmax'],
        'class_1_ymax': data['class_1_coords']['ymax'],
        'class_0_xmin': data['class_0_coords']['xmin'],
        'class_0_ymin': data['class_0_coords']['ymin'],
        'class_0_xmax': data['class_0_coords']['xmax'],
        'class_0_ymax': data['class_0_coords']['ymax'],
        'iou': iou,
        'above': is_above
    })

summary_df = pd.DataFrame(summary_data)
print("\nSummary DataFrame:")
print(summary_df)
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np
import pandas as pd
import pickle
from time import time
from tqdm import tqdm
from PIL import Image
import torch
import mmcv
from torchvision.ops import nms




iou_thr = 0.4
conf_score_class_0 = 0.3  
conf_score_class_1_initial = 0.15
conf_score_class_1_min = 0.001  
conf_step = 0.01  

device = "cuda:0" if torch.cuda.is_available() else "cpu"
print(f"Using device: {device}")





current_predictions = submission_df.copy()



def has_class_1_in_predictions(pred_string):
    """Check if prediction string contains any class 1 detections"""
    if pred_string == '-':
        return False
    
    detections = pred_string.split(' 0 ')  
    for det in detections:
        if det.strip().startswith('1 '):
            return True
    return False




def parse_predictions(pred_string):
    """Parse prediction string into list of (class, conf, x, y, w, h)"""
    if pred_string == '-':
        return []
    
    parts = pred_string.split()
    predictions = []
    i = 0
    while i < len(parts):
        if i + 5 < len(parts):
            cls = int(parts[i])
            conf = float(parts[i + 1])
            x = float(parts[i + 2])
            y = float(parts[i + 3])
            w = float(parts[i + 4])
            h = float(parts[i + 5])
            predictions.append((cls, conf, x, y, w, h))
            i += 6
        else:
            break
    return predictions




def format_predictions(predictions):
    """Format list of predictions back to string"""
    if not predictions:
        return '-'
    
    parts = []
    for cls, conf, x, y, w, h in predictions:
        parts.append(f"{cls} {conf:.6f} {x:.6f} {y:.6f} {w:.6f} {h:.6f}")
    return ' '.join(parts)




def run_inference_with_threshold(model, img_path, conf_class_1):
    """Run inference with specific class 1 threshold"""
    img = mmcv.imread(img_path)
    height, width = img.shape[:2]
    
    
    det_result = inference_detector(model, img)
    pred_instances = det_result.pred_instances
    
    bboxes = pred_instances.bboxes
    scores = pred_instances.scores
    labels = pred_instances.labels
    
    
    class_0_mask = (labels == 0) & (scores >= conf_score_class_0)
    class_1_mask = (labels == 1) & (scores >= conf_class_1)
    
    high_conf_indices = class_0_mask | class_1_mask
    
    bboxes = bboxes[high_conf_indices]
    scores = scores[high_conf_indices]
    labels = labels[high_conf_indices]
    
    
    if len(bboxes) > 0:
        keep_indices = nms(bboxes, scores, iou_thr)
        bboxes = bboxes[keep_indices].cpu().numpy()
        scores = scores[keep_indices].cpu().numpy()
        labels = labels[keep_indices].cpu().numpy()
        
        
        predictions = []
        for bbox, score, label in zip(bboxes, scores, labels):
            xmin, ymin, xmax, ymax = bbox
            
            
            xmin_norm = xmin / width
            ymin_norm = ymin / height
            xmax_norm = xmax / width
            ymax_norm = ymax / height
            
            
            x_center = (xmin_norm + xmax_norm) / 2
            y_center = (ymin_norm + ymax_norm) / 2
            w = xmax_norm - xmin_norm
            h = ymax_norm - ymin_norm
            
            predictions.append((int(label), float(score), x_center, y_center, w, h))
        
        return predictions
    
    return []





problematic_cases = current_predictions[
    (current_predictions['right_place'] == 1) & 
    (current_predictions['has_class_1'] == 0)
].copy()

print(f"Found {len(problematic_cases)} cases that need class 1 detections")
print(f"Problematic files: {problematic_cases['filename'].tolist()}")


updated_predictions = current_predictions.copy()

for idx, row in tqdm(problematic_cases.iterrows(), total=len(problematic_cases), desc="Fixing class 1 detections"):
    filename = row['filename']
    current_preds = parse_predictions(row['prediction_string'])
    
    
    image_id = filename.replace('.jpg', '')
    img_path = f'/kaggle/input/datasets/kagglertw/dal-shemagh/dal-shemagh-detection-challenge/images/test/{image_id}.jpg'
    
    print(f"\nProcessing {filename}...")
    
    
    conf_threshold = conf_score_class_1_initial
    found_class_1 = False
    
    while conf_threshold >= conf_score_class_1_min and not found_class_1:
        conf_threshold -= conf_step
        
        
        new_predictions = run_inference_with_threshold(mmdet_model_list[0], img_path, conf_threshold)
        
        
        class_1_preds = [p for p in new_predictions if p[0] == 1]
        
        if class_1_preds:
            found_class_1 = True
            print(f"  Found class 1 at threshold {conf_threshold:.3f}")
            
            
            best_class_1 = max(class_1_preds, key=lambda x: x[1])
            
            
            combined_preds = current_preds + [best_class_1]
            
            
            new_pred_string = format_predictions(combined_preds)
            
            
            updated_predictions.loc[idx, 'prediction_string'] = new_pred_string
            updated_predictions.loc[idx, 'has_class_1'] = 1
            
            print(f"  Added: class {best_class_1[0]}, conf {best_class_1[1]:.4f}")
    
    if not found_class_1:
        print(f"  WARNING: Could not find class 1 even at threshold {conf_score_class_1_min}")




print("\n" + "="*60)
print("SUMMARY")
print("="*60)
print(f"Original cases without class 1: {len(problematic_cases)}")
print(f"Fixed cases: {updated_predictions['has_class_1'].sum() - current_predictions['has_class_1'].sum()}")
print(f"Remaining issues: {len(updated_predictions[(updated_predictions['right_place'] == 1) & (updated_predictions['has_class_1'] == 0)])}")


output_path = '/kaggle/working/updated_predictions.csv'
updated_predictions.to_csv(output_path, index=False)
print(f"\nSaved updated predictions to: {output_path}")


print("\n" + "="*60)
print("BEFORE vs AFTER (for problematic cases)")
print("="*60)
for idx, row in problematic_cases.iterrows():
    filename = row['filename']
    old_pred = row['prediction_string']
    new_pred = updated_predictions.loc[idx, 'prediction_string']
    
    if old_pred != new_pred:
        print(f"\n{filename}:")
        print(f"  OLD: {old_pred}")
        print(f"  NEW: {new_pred}")
```
