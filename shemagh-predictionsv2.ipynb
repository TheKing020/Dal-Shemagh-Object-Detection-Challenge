```python
rp = pd.read_csv('/kaggle/input/notebooks/kagglertw/shemagh-right-place-binary-inferv2/convnext_predictions.csv')
```

```python
rp
```

```python
rp['prediction_class'] = (rp[['fold_1_prob', 'fold_2_prob', 'fold_3_prob', 'fold_4_prob', 'fold_5_prob']].mean(axis=1) > 0.63).astype(int)
```

```python
for _, row in rp.iterrows():
    sub.loc[
        (sub['filename'] == row['filename']) & (sub['right_place'] == 0),
        'right_place'
    ] = row['prediction_class']
```

```python
sub.to_csv('testing_.csv' , index = False)
```
